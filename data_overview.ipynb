```python
import json
import torch
import numpy as np
```

```python
train_path = 'drive/MyDrive/DeepLearningProject_Group3/preprocessed/data_preprocessed/train.json'
dev_path = 'drive/MyDrive/DeepLearningProject_Group3/preprocessed/data_preprocessed/dev.json'
test_path = 'drive/MyDrive/DeepLearningProject_Group3/preprocessed/data_preprocessed/test.json'
```

# Number of samples per classes

```python
with open(train_path, 'r') as file:
    train_data = json.load(file)
```

```python
import seaborn as sns

y_train = train_data["sentiments"]
print(len(y_train))
samples_per_classes = [y_train.count(0), y_train.count(1), y_train.count(2)]
sns.barplot(x=list(range(3)), y=samples_per_classes)
```

```python
with open(dev_path, 'r') as file:
    dev = json.load(file)
```

```python
y_dev = dev["sentiments"]
print(len(y_dev))
samples_per_classes = [y_dev.count(0), y_dev.count(1), y_dev.count(2)]
sns.barplot(x=list(range(3)), y=samples_per_classes)
```

```python
with open(test_path, 'r') as file:
    test = json.load(file)
```

```python
y_test = test["sentiments"]
print(len(y_test))
samples_per_classes = [y_test.count(0), y_test.count(1), y_test.count(2)]
sns.barplot(x=list(range(3)), y=samples_per_classes)
```

# Number of samples per topics

```python
y_train_topic = train_data["topics"]
samples_per_topics = [y_train_topic.count(i) for i in range(4)]
sns.barplot(x=["Lecturer", "Program", "Facility", "Others"], y=samples_per_topics)
```

```python
y_dev_topic = dev["topics"]
samples_per_topics = [y_dev_topic.count(i) for i in range(4)]
sns.barplot(x=["Lecturer", "Program", "Facility", "Others"], y=samples_per_topics)
```

# Sequence Length Distribution

```python
import matplotlib.pyplot as plt

x_train = train_data["word_embedding"]
sentence_length = [len(sentence) for sentence in x_train]

plt.hist(sentence_length, bins=50)
plt.xlabel("Sentence Length")
plt.title("Distribution of Sentence Length")
plt.show()
```
